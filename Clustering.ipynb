## HIERARCHY CLUSTERING

```python
from scipy.cluster.hierarchy import dendrogram,linkage
from  scipy.spatial.distance import pdist
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
xdist=pdist(iris,metric="euclidean")
linked=linkage(xdist,'ward')


plt.figure(figsize=(10,7))
dendrogram(linked,orientation="top",distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage='ward')
cluster.fit_predict(irisS)
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity="l2", linkage='average')
cluster.fit_predict(irisS)
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity="l1", linkage='single')
cluster.fit_predict(irisS)
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity="manhattan", linkage='complete')
cluster.fit_predict(irisS)
```

## DB SCAN

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(100,1)
```

```python
data=pd.DataFrame(x)
```

```python
data.columns=["f1","f2"]
```

```python
data["class"]=y
```

```python
data.head()
```

```python
plt.scatter(data[data["class"]==1].iloc[:,0],data[data["class"]==1].iloc[:,1])
plt.scatter(data[data["class"]==0].iloc[:,0],data[data["class"]==0].iloc[:,1])
```

```python
xdist=pdist(x,metric="euclidean")
linked=linkage(xdist,'ward')
plt.figure(figsize=(10,7))
dendrogram(linked,orientation="top",distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
iris = pd.DataFrame(x)
scaler = MinMaxScaler()
irisS = scaler.fit_transform(iris.iloc[:,:-1])
model = KMeans(n_clusters=3,max_iter=600)
model.fit(irisS)
pred = model.predict(irisS)
plt.scatter(iris.iloc[:,0],iris.iloc[:,1],c=pred)
```

> ***************************************
> Selecting Optimal no. of Clusters
> ***************************************

* Distances between clusters should be Heterogeneous
* Distances between points in cluster should be Homogeneous

```python
iris = sns.load_dataset('iris')
```

```python
nc = np.arange(2,10)
```

```python
iner = []
for i in nc :
    km = KMeans(i)
    km.fit(iris.iloc[:,:4])
    iner.append(km.inertia_)
```

```python
iner
```

```python
plt.plot(nc,iner)
plt.scatter(nc,iner,color='r',marker='s')
plt.show()
```

```python
ld = pd.read_csv('diabetes.csv')
```

```python
iner = []
for i in nc :
    km = KMeans(i)
    km.fit(ld.iloc[:,:-1])
    iner.append(km.inertia_)
```

```python
plt.figure(figsize=(16,9))
plt.plot(nc,iner)
plt.scatter(nc,iner,color='g',marker='X')
plt.title('Optimal "k" Value for Kmeans')
plt.show()
```
